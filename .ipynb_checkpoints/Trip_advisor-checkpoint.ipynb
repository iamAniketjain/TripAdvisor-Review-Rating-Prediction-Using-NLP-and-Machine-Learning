# TripAdvisor Review Rating Prediction Using NLP and Machine Learning

# Aim
* To analyze TripAdvisor hotel review text using NLP techniques.

* To convert raw text data into meaningful numerical features using vectorization methods.

* To build a machine learning model that predicts the star rating (1–5) based on the review text.

* To create an automated system that can classify reviews without manual reading.

* To support businesses in identifying customer satisfaction levels using AI-driven insights.

# The dataset consists of 20491 rows and 2 column variables. Each row represents a hotel review submitted by a customer on TripAdvisor. The dataset includes the textual review written by the user and the corresponding numerical rating assigned. This data is suitable for Natural Language Processing (NLP) tasks such as sentiment analysis, text classification, and rating prediction.

# Feature Information

* 1 – Review :
Text data containing customer-written hotel reviews from TripAdvisor. These reviews describe the user's experience, opinions, feedback, and sentiments regarding their stay.

* 2 – Rating :
Numerical category representing the star rating given by the reviewer for the hotel. Rating values typically range from 1 to 5.

# Punctuation

```python
string.punctuation
```

```python
def text_preocess(text):
    """
    remove punctuations
    remove stopwords
    returns the list of text words
    """
    nopunc =[char for char in text if char not in string.punctuation]
    nopunc = "".join(nopunc)
    return [word for word in nopunc.split() if word not in l1]
```

```python
trip['x'].apply(text_preocess)
```

# Pre Processing Data

# Split the Data into Train-Test 

```python
x_train, x_test, train_y, test_y = train_test_split(trip['x'], trip['y'], random_state=42, test_size=.2)
train_y.shape
```

```python
trip.y.value_counts()
```

# TF-IDF

```python
tfidf =TfidfVectorizer(ngram_range=(1,2),
                        min_df=2,
                        max_df=0.95,
                      stop_words='english')
x_train_tfidf = tfidf.fit_transform(x_train)
x_test_tfidf = tfidf.transform(x_test)
```

```python
ros = RandomOverSampler()
```

```python
x_train_ros, y_train_ros = ros.fit_resample(x_train_tfidf, train_y)
```

# Naive Bayes

```python
nb = MultinomialNB(alpha=0.3)
```

```python
nb.fit(x_train_ros, y_train_ros)
```

```python
for a in [0.01, 0.1, 0.3, 0.5, 1]:
    print(a, MultinomialNB(alpha=a).fit(x_train_ros, y_train_ros).score(x_train_ros, y_train_ros))
```

```python
pred = nb.predict(x_test_tfidf)
```

```python
tab = confusion_matrix(test_y, pred)
tab
```

```python
accuracy_score(test_y, pred)
```

```python
for a in [0.01, 0.1, 0.3, 0.5, 1]:
    nb = MultinomialNB(alpha=a)
    nb.fit(x_train_ros, y_train_ros)
    
    train_acc = nb.score(x_train_ros, y_train_ros)
    test_acc  = nb.score(x_test_tfidf, test_y)

    print(f"alpha={a}, train={train_acc:.4f}, test={test_acc:.4f}")
```

# LinearSVM

```python
lsvm = LinearSVC()
```

```python
lsvm.fit(x_train_ros, y_train_ros)
```

```python
predsvm= lsvm.predict(x_test_tfidf)
```

```python
tabsvm = confusion_matrix(test_y, predsvm)
tabsvm
```

```python
accuracy_score(test_y, predsvm)
```

# Logistic Regression

```python
lr = LogisticRegression(max_iter=3000)
lr.fit(x_train_ros, y_train_ros)
```

```python
predls = lr.predict(x_test_tfidf)
```

```python
accuracy_score(test_y,predls)
```

# Lower String

```python
new_trip.x.str.lower()
```

```python
l1 = stopwords.words('english')
```

```python
string.punctuation
```

```python
def text_convert(data):
    nopunc=[char for char in data if char not in string.punctuation]
    nopunc = "".join(nopunc)
    return [word for word in nopunc.split() if word not in l1]
```

```python
new_trip['x'].apply(text_convert)
```

# Create TDM

```python
start = timeit.default_timer()
bow_transformer = CountVectorizer(analyzer=text_convert).fit(new_trip['x'])
stop = timeit.default_timer()
execution_time = stop-start
print("execution complete in ",execution_time)
```

```python
bow_transformer = CountVectorizer(analyzer=text_convert).fit(new_trip['x'])
```

```python
bow_transformer.vocabulary_
```

```python
tdm = bow_transformer.transform(new_trip['x'])
```

# Split the Data into train-test
